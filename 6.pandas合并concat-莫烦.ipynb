```python
import numpy as np
import pandas as pd
```

# concatenating

```python
df1 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2,columns=['a','b','c','d'])
```

```python
df1
```

```python
df2
```

```python
df3
```

# 这里属性是相同的，所以属于上下合并

```python
res = pd.concat([df1, df2, df3], axis=0)
```

```python
res
```

# 重新排序 ignore_index=True

```python
res = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
```

```python
res
```

# join

```python
df1 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'], index=[1,2,3])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['b','c','d','e'], index=[2,3,4])
```

```python
df1
```

```python
df2
```

# outer--default（并集）

```python
res = pd.concat([df1,df2], join='outer')
res
```

# inner--只保留共同部分（交集）

```python
res = pd.concat([df1,df2], join='inner',ignore_index=True)
res
```

# reindex

```python
df1 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'], index=[1,2,3])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['b','c','d','e'], index=[2,3,4])
```

```python
# 按照df1的index进行合并，df2中没有对应index的用NaN填充，不属于index的直接丢弃
res = pd.concat([df1,df2],axis=1).reindex(df1.index)
res
```

# append

```python
df1 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*1,columns=['b','c','d','e'],index=[2,3,4])
res=df1.append(df2, ignore_index=True)
res
```

```python
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'])
s1 = pd.Series([1,2,3,4], index=['a','b','c','d'])
res = df1.append(s1, ignore_index=True)
res
```

# 注意下面这里的.to_frame().T

```python
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'])
s1 = pd.Series([1,2,3,4], index=['a','b','c','d'])
res = pd.concat([df1, s1.to_frame().T], ignore_index=True)
res
```
