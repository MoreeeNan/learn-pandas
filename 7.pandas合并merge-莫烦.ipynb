```python
import numpy as np
import pandas as pd
```

```python
left = pd.DataFrame({
    'key': ['K0','K1','K2','K3'],
    'A': ['A0','A1','A2','A3'],
    'B': ['B0','B1','B2','B3']
})
left
```

```python
right = pd.DataFrame({
    'key': ['K0','K1','K2','K3'],
    'C': ['C0','C1','C2','C3'],
    'D': ['D0','D1','D2','D3']
})
right
```

# merge two df by key/keys (may be used in database)

# one key

```python
res = pd.merge(left, right, on='key')
res
```

# consider two keys

```python
left = pd.DataFrame({
    'key1': ['K0','K0','K1','K2'],
    'key2': ['K0','K1','K0','K1'],
    'A': ['A0','A1','A2','A3'],
    'B': ['B0','B1','B2','B3']
})
left
```

```python
right = pd.DataFrame({
    'key1': ['K0','K1','K1','K2'],
    'key2': ['K0','K0','K0','K0'],
    'C': ['C0','C1','C2','C3'],
    'D': ['D0','D1','D2','D3']
})
right
```

# how = ['left','right','inner','outer']

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='inner')   # how='inner'为default
res
```

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='outer')   
res
```

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='right')   # 按照右边的df，也就是第二个传入的df进行merge
res
```

# indicator

```python
df1 = pd.DataFrame({
    'col1':[0,1],
    'col_left':['a','b']
})
df1
```

```python
df2 = pd.DataFrame({
    'col1': [1,2,2],
    'col_right': [2,2,2]
})
df2
```

```python
res = pd.merge(df1, df2, on='col1', how='outer', indicator=True)
res
```

# 给indicator一个指定的名字

```python
res = pd.merge(df1,df2, on='col1',how='outer',indicator='indicator_column')
res
```

# merge by index

```python
left = pd.DataFrame(
    {
    'A': ['A0','A1','A2'],
    'B': ['B0','B1','B2']
    },
    index=['K0','K1','K2']
)
left
```

```python
right = pd.DataFrame(
    {
    'C': ['C0','C1','C2'],
    'D': ['D0','D1','D2']
    },
    index=['K0','K2','K3']
)
right
```

```python
res = pd.merge(left, right, left_index=True, right_index=True, how='outer')
res
```

# handle overlapping

```python
# 例如这里的boy和girl都有age属性，但在合并的时候必须得区分开来成age_boy和age_girl
```

```python
boys = pd.DataFrame(
    {
        'k':['K0','K1','K2'],
        'age':[1,2,3]
    }
)
boys
```

```python
girls = pd.DataFrame({
    'k':['K0','K0','K3'],
    'age':[4,5,6]
})
girls
```

```python
res = pd.merge(boys, girls, on='k',suffixes=['_boy','_girl'], how='inner')
res
```

```python
res = pd.merge(boys, girls, on='k',suffixes=['_boy','_girl'], how='outer')
res
```
